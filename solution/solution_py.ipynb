```python
# Creates a varaible name for each medication file in folder [medication1, medication2, ... ,medicationx]

# Note that variables 0 to 14 are now mapped to filenames 1, 10, 11, 12, 13, 14, 15, 2, 3, 4, 5, 6, 7, 8, 9 respectively

medication_list = glob.glob(r'../exa-data-analyst-assessment/data/medication/*.csv')
for n, val in enumerate(medication_list):
    globals()["medication%d"%n] = pd.read_csv(val)

# Creates a variable name for each observation file in folder [observation1, observation2, ..., observationx]

# Note that variables 0 to 11 are now mapped to filenames 1, 10, 11, 12, 2, 3, 4, 5, 6, 7, 8, 9 respectively

observation_list = glob.glob(r'../exa-data-analyst-assessment/data/observation/*.csv')
for n, val in enumerate(observation_list):
    globals()["observation%d"%n] = pd.read_csv(val)
```
